```python
# So, basically I have been given a data set of legos, one of my favourite childhood game. The data of legos is scraped from 
# data camp. So our task is to get some insight of the data and answer some of the problems. 
# Task 1: The organization needs to know that how much percentage of legos data set are "Star Wars" and are "licensed". 
# Task 2: In which year, the maximum number of licensed "Star Wars" nunber sets were produced. 
# Task 3: In which year, the maximum number of the unlicensed sets were produced. 

import pandas as pd

# reading the data set 
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
# joining the two data sets i.e. parent_theme and df by inner join by parent_theme(column1) = name(coulmn2)
merged = df.merge(parent_theme, left_on = 'parent_theme', right_on = 'name')
```

```python
# sorting the values by year. 
licensed_sorted = licensed.sort_values('year')
licensed_sorted
```

```python
# Making a new column name count with all values as 1. 
licensed_sorted['count'] = 1
```

```python
# group by year and then parent_theme, sum all the value in each group, by reset_index() have the new indexed with 0 starting. 
summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()
```

```python
# Printing the head of the data frame name summed_df. 
summed_df.head()
```

```python
# now, sorting the values by count in descending order, now we can drop the rest of the duplicates, after maximum count. 
max_df = summed_df.sort_values('count', ascending = False).drop_duplicates(['year'])
```

```python
# Print the new data frame max_df. 
max_df.shape
max_df.head(19)
```

```python
# By seeing through the data set, year 2017 has the minimum number of licensed "Star Wars" 
# Your task 2 is completed 
new_era = 2017
```

```python
# Task 3: In which year maximum number of sets were produced with no license 
clean_df = merged[~merged['set_num'].isnull()]
```

```python
# This is the clean data frame with no NaN values. 
clean_df
```

```python
# Adding a new column in clean_df i.e. countit = 1 
clean_df['countit'] = 1
```

```python
# Prints the data frame with new column added all values to 1. 
clean_df
```

```python
# For each year, sum the values for each column by group and create a new data frame name sets_per_year. 
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year', 'countit']]
```

```python
# prints the year which has the maximum number of sets per year. 
max_sets_year = sets_per_year[sets_per_year['countit'] == sets_per_year['countit'].max()]['year']
```

```python
# The answer to task 3 is 2014. 
print(max_sets_year)
```
